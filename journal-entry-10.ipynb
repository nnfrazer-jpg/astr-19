```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.standard_exponential(1000)
plt.xlabel('x')
plt.ylabel('y')
plt.hist(x,bins=100,alpha=0.5,edgecolor="black")
plt.savefig('entry-10-histogram.pdf',bbox_inches='tight')
```
